```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from geopy.distance import great_circle

# Sample data: locations of delivery points (latitude, longitude)
delivery_points = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'latitude': [12.9716, 13.0827, 12.2958, 13.0674],
    'longitude': [77.5946, 80.2707, 76.6393, 80.2376]
})

# Clustering delivery points to optimize routes
kmeans = KMeans(n_clusters=2)  # Number of clusters can be adjusted
delivery_points['cluster'] = kmeans.fit_predict(delivery_points[['latitude', 'longitude']])

# Calculate distances between points in each cluster
def calculate_distances(cluster_points):
    distances = []
    for i in range(len(cluster_points)):
        for j in range(i + 1, len(cluster_points)):
            dist = great_circle(
                (cluster_points.iloc[i]['latitude'], cluster_points.iloc[i]['longitude']),
                (cluster_points.iloc[j]['latitude'], cluster_points.iloc[j]['longitude'])
            ).kilometers
            distances.append((cluster_points.iloc[i]['id'], cluster_points.iloc[j]['id'], dist))
    return distances

# Example usage
for cluster in delivery_points['cluster'].unique():
    cluster_points = delivery_points[delivery_points['cluster'] == cluster]
    distances = calculate_distances(cluster_points)
    print(f"Distances for Cluster {cluster}: {distances}")
```

```python
import matplotlib.pyplot as plt

# Sample data: emissions over time
months = ['January', 'February', 'March', 'April']
emissions = [1200, 1150, 1100, 1000]  # Example emission data in kg CO2

plt.plot(months, emissions, marker='o')
plt.title('Monthly Emissions Reduction')
plt.xlabel('Month')
plt.ylabel('Emissions (kg CO2)')
plt.grid()
plt.show()
```

```python
packaging_materials = []

def add_packaging(material, recyclable):
    packaging_materials.append({'material': material, 'recyclable': recyclable})

def list_packaging():
    return pd.DataFrame(packaging_materials)

# Adding some packaging materials
add_packaging('Biodegradable Plastic', True)
add_packaging('Recycled Cardboard', True)
add_packaging('Styrofoam', False)

# Listing all packaging materials
packaging_df = list_packaging()
print(packaging_df)
```

```python
training_programs = []

def add_training(name):
    training_programs.append({'name': name})

def list_trainings():
    return pd.DataFrame(training_programs)

# Adding training programs
add_training('Sustainability Awareness')
add_training('Eco-Friendly Packaging Practices')

# Listing all training programs
trainings_df = list_trainings()
print(trainings_df)
```
